```python
from nltk.stem import PorterStemmer
from yellowbrick.cluster import KElbowVisualizer
def StemWords(word_list):
    porter = PorterStemmer()
    new_word_list = ""
    for word in word_list.split(' '):
        new_word_list += porter.stem(word) + ' ' 
        
    return new_word_list
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer

def StemAndLem(word_list):  
    porter = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    new_word_list = ""
    for word in word_list.split(' '):
        if len(word) > 3:
            new_word_list += porter.stem(lemmatizer.lemmatize(word)) + ' ' 
        
    return new_word_list
```

```python
from nltk.corpus import stopwords
import re
from nltk.tokenize import word_tokenize


def removestopwords(word_list):
    pattern = re.compile(r'\b(' + r'|'.join(stopwords.words('english')) + r')\b\s*')
    return pattern.sub('', word_list)
```

```python
from sklearn.neighbors import NearestNeighbors

def find_k_closest(centroids, data, num):
    nns = {}
    neighbors = NearestNeighbors(n_neighbors=num).fit(data)
    for center in centroids:
        nns[center] = neighbors.kneighbors(center, return_distance=False)
    return nns
```

```python
import csv
import os
import json
from collections import defaultdict
import nltk
import datetime
from sklearn.neighbors import NearestNeighbors
from nltk.corpus import stopwords
import re

def GetArticleFromDate(startDate, endDate):

    cord_uid_to_text = defaultdict(list)


    with open('metadata.csv') as f_in:
        reader = csv.DictReader(f_in)
        for row in reader:           
            if '-' not in row['publish_time']:
                continue
            elif startDate > datetime.datetime.strptime(row['publish_time'], '%Y-%m-%d').date() or datetime.datetime.strptime(row['publish_time'], '%Y-%m-%d').date() > endDate:
                continue
            cord_uid = row['cord_uid']
            title = row['title']
            abstract = row['abstract']
            authors = row['authors'].split('; ')

            introduction = []
            if row['pmc_json_files']:
                for json_path in row['pmc_json_files'].split('; '):
                    with open(json_path) as f_json:
                        full_text_dict = json.load(f_json)
                    
                        for paragraph_dict in full_text_dict['body_text']:
                            paragraph_text = paragraph_dict['text']
                            section_name = paragraph_dict['section']
                            introduction.append(paragraph_text)

            cord_uid_to_text[cord_uid].append(introduction)
    
  
    cleanedUpDocList = []
    originalDocList= []
    for key, value in cord_uid_to_text.items():
        doc = ''
        doc2 = ''
        for v in value:
            for m in v:
                if m:
                    cleanedUp = re.sub(r'\[(.*)]*\]','',re.sub(r'\([^)]*\)', '', m))
                    doc2 += re.sub(r'\.+', ".", cleanedUp) 
                    doc += re.sub(r'\.+', ".", StemAndLem(removestopwords(re.sub (r'([^a-zA-Z \.]+?)', '', cleanedUp.lower()))))
                  
        if doc and doc.count('.') > 4:
            cleanedUpDocList.append(doc)
            originalDocList.append(doc2)
    
    print(len(cleanedUpDocList))
    
    return cleanedUpDocList, originalDocList

    
```

```python

from sklearn.decomposition import TruncatedSVD

start = datetime.date(2020, 3, 1)
end = datetime.date(2020, 3, 7)

corpus, og = GetArticleFromDate(start, end)
corpus = og
sentences = []
for corpi in corpus:    
    for sentence in sent_tokenize(corpi):
        sentences.append(sentence)
    
print(len(sentences))
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

svd = TruncatedSVD(n_components=100)
svd.fit(X)

X = svd.transform(X)

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20), timings=True)

visualizer.fit(X)        # Fit the data to the visualizer

clustering_model = KMeans(n_clusters=visualizer.elbow_value_).fit(X)

half = int((len(sentences)/visualizer.elbow_value_)*.8)
print(half)
neigh = NearestNeighbors(n_neighbors=half, radius=0.4)
neigh.fit(X)
print(datetime.datetime.now())

ind = neigh.kneighbors(clustering_model.cluster_centers_, int(neigh.n_samples_fit_*.8), return_distance=False)

sentencesToRemove = []
for index in ind:
    for i in index:
        if len(sentences[i].split(' ')) > 5:
            sentencesToRemove.append(sentences[i])
        #print(sentences[i])
print(len(sentencesToRemove))


newCorpus = []
for corpi in corpus:    
    newCorpiSentences = []
    for sentence in sent_tokenize(corpi):
        if sentence in sentencesToRemove:
            #print(sentence)
            newCorpiSentences.append(sentence)
                
    newCorpus.append(' '.join(newCorpiSentences))
        
sentences2 = []
for corpi in newCorpus:    
    for s in sent_tokenize(corpi):
        sentences2.append(s)
            
print(len(sentences2))

cleanedUpDocList = []
for value in newCorpus:
    doc = ''
    doc2 = ''
    if value:
        cleanedUp = re.sub(r'\[(.*)]*\]','',re.sub(r'\([^)]*\)', '', value))
        doc2 += re.sub(r'\.+', ".", cleanedUp) 
        doc += re.sub(r'\.+', ".", StemAndLem(removestopwords(re.sub (r'([^a-zA-Z \.]+?)', '', cleanedUp.lower()))))
                  
        if doc and doc.count('.') > 1:
            cleanedUpDocList.append(doc)
        
corpus = cleanedUpDocList

print('Test 4, June-July, with clustering preprocessing')

print('TF-IDF Test')
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

svd = TruncatedSVD(n_components=100)
svd.fit(X)

X_pca = svd.transform(X)
print(datetime.datetime.now())

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10), timings=False, locate_elbow=False)

visualizer.ax.set_xlabel("Sum of Squared Distances")
visualizer.fit(X_pca)       
visualizer.show()     
print(datetime.datetime.now())
print('bert Test')
embedder = SentenceTransformer('google/bert_uncased_L-12_H-512_A-8')
X = embedder.encode(corpus)

svd = TruncatedSVD(n_components=100)
svd.fit(X)

X_pca = svd.transform(X)

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20), timings=False, locate_elbow=False)

visualizer.ax.set_xlabel("Sum of Squared Distances")
visualizer.fit(X_pca)       
visualizer.show() 

print(datetime.datetime.now())

print('distilbert Test')
embedder = SentenceTransformer('distilbert-base-uncased-finetuned-sst-2-english')
X = embedder.encode(corpus)

print(datetime.datetime.now())
svd = TruncatedSVD(n_components=100)
svd.fit(X)

X_pca = svd.transform(X)
print(datetime.datetime.now())
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20), timings=False, locate_elbow=False)
print(datetime.datetime.now())
visualizer.ax.set_xlabel("Sum of Squared Distances")
visualizer.fit(X_pca)       
visualizer.show()   
print(datetime.datetime.now())

print('Roberta Test')
embedder = SentenceTransformer('sentence-transformers/msmarco-roberta-base-v2')
X = embedder.encode(corpus)

svd = TruncatedSVD(n_components=100)
svd.fit(X)

X_pca = svd.transform(X)

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20), timings=False, locate_elbow=False)

visualizer.ax.set_xlabel("Sum of Squared Distances")
visualizer.fit(X_pca)       
visualizer.show() 


print(datetime.datetime.now())

print('Glove Test')
embedder = SentenceTransformer('sentence-transformers/average_word_embeddings_glove.6B.300d')
X = embedder.encode(corpus)

svd = TruncatedSVD(n_components=100)
svd.fit(X)

X_pca = svd.transform(X)

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20), timings=False, locate_elbow=False)

visualizer.ax.set_xlabel("Sum of Squared Distances")
visualizer.fit(X_pca)       
visualizer.show() 

print(datetime.datetime.now())

print('Labse Test')
embedder = SentenceTransformer('sentence-transformers/LaBSE')
X = embedder.encode(corpus)

svd = TruncatedSVD(n_components=100)
svd.fit(X)

X_pca = svd.transform(X)

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,20), timings=False, locate_elbow=False)

visualizer.ax.set_xlabel("Sum of Squared Distances")
visualizer.fit(X_pca)       
visualizer.show()  
print(datetime.datetime.now())
```
